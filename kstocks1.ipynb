```python
import requests
import io
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
stock = "P"
url = "http://ichart.finance.yahoo.com/table.csv?s=" + stock + "&d=0&e=1&f=2017&g=d&a=0&b=1&c=2016&ignore=.csv"
```

```python
s = requests.get(url).content
```

```python
c = pd.read_csv(io.StringIO(s.decode("utf-8")))
```

```python
c["mydate"]= pd.to_datetime(c["Date"])
```

```python
c["diff"]=c["Close"]-c["Open"]
```

```python
c.head()
```

```python
c.plot(x=7,y=1,kind="line",figsize=[10,5])
```
