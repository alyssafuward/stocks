```python
import requests
import io
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
%matplotlib inline
```

```python
url = "http://ichart.finance.yahoo.com/table.csv?s=TSLA&d=0&e=1&f=2017&g=d&a=0&b=1&c=2016&ignore=.csv"
```

```python
s = requests.get(url).content
```

```python
c = pd.read_csv(io.StringIO(s.decode("utf-8")))
```

```python
c.head()
```

```python
c["Date"] = pd.to_datetime(c["Date"])
```

```python
def graph_stuff(x_var,y_var):
    plt.figure()
    plt.plot(x_var,y_var)
    plt.show()
```

```python
for column in ["High", "Low","Open","Close","Volume"]:
    print(column)
    graph_stuff(c["Date"],c[column])
```
